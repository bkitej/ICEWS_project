```python
import pandas as pd
import zipfile, io, os
from tqdm import tqdm
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cd icews_files
```

```python
# reference: https://stackoverflow.com/questions/31346790/unzip-all-zipped-files-in-a-folder-to-that-same-folder-using-python-2-7-5

if 0:
    for item in tqdm(os.listdir('.')): # loop through items in dir
        try:
            if not item.endswith('.zip'):
                pass
            else:
                file_name = os.path.abspath(item)    # get full path of files
                zip_ref = zipfile.ZipFile(file_name) # create zipfile object
                zip_ref.extractall('.')              # extract file to dir
                zip_ref.close()                      # close file
                os.remove(file_name)                 # delete zipped file
                print(item + ' unzipped successfully.')
        except zipfile.BadZipFile:
            print(item + ' could not be unzipped.')
```

```python
def read_icews(filename):
    
    "Read single ICEWS .tab file into a DataFrame."
    
    try:     
        # in case there are non-tab files in the directory
        # return empty DataFrame so as not to break pd.concat()
        if not filename.endswith('.tab'):
            return pd.DataFrame()
        
        #
        else:
            df = pd.read_csv(filename, sep = '\t')
            return df
        
    # should be pd.errors.ParseError
    except:
        print(filename + ' could not be read.')
        return pd.DataFrame()
```

```python
# generator of files in icews_files directory
dfs = (read_icews(filename) for filename in tqdm(os.listdir('.')))
```

```python
# stack DataFrames
df = pd.concat(dfs, ignore_index = True)
```

```python
# convert to datetime
df['Event Date'] = pd.to_datetime(df['Event Date'])
```

```python
# sort
df = df.sort_values('Event Date')
```

```python
by_date = df.groupby('Event Date').apply(len)
```

```python
weekdays = by_date[by_date.index.weekday <= 4]
weekends = by_date[by_date.index.weekday > 4]
```

```python
weekdays_mean = weekdays.mean()
weekends_mean = weekends.mean()
```

```python
sns.set_style('whitegrid')
fig, ax = plt.subplots(1,1, figsize = (16,10))
ax = sns.lineplot(x = by_date.index, y = by_date.values)
ax.set_title('ICEWS Events Over Time', fontsize = 40);
```
